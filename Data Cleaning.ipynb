```python
#2.Products
products=pd.read_csv("C:/Users/prem/Desktop/Ranjana/2.Data Spark Project/Products.csv")
products
```

```python
products.info()
```

```python
# Remove non-numeric characters like $ and ,
products['Unit Cost USD'] = products['Unit Cost USD'].replace('[\$,]', '', regex=True).astype(float)
products['Unit Price USD'] = products['Unit Price USD'].replace('[\$,]', '', regex=True).astype(float)
```

```python
#ceck any missing values
missing_values = products.isnull().sum()
print(missing_values)
```

```python
#checking for unique combinations
unique_combinations = products.groupby(['SubcategoryKey', 'Subcategory']).size()
print(unique_combinations)
```

```python
#to ensure that Unit Price USD is always greater than or equal to Unit Cost USD to avoid data errors
inconsistent_prices = products[products['Unit Price USD'] < products['Unit Cost USD']]
print(inconsistent_prices)
```

```python
#create a new data set
products.to_csv("C:/Users/prem/Desktop/Ranjana/2.Data Spark Project/New Data Set/Products.csv", index=False)
```

```python
#3.stores
stores=pd.read_csv("C:/Users/prem/Desktop/Ranjana/2.Data Spark Project/Stores.csv")
stores
```

```python
stores.info()
```

```python
# Fill missing values with the mean
stores['Square Meters'] = stores['Square Meters'].fillna(stores['Square Meters'].mean())
```

```python
stores['Open Date']= stores['Open Date'].astype(str).str.replace("-","/")
```

```python
#date convert
stores['Open Date'] = pd.to_datetime(stores['Open Date'], format='%m/%d/%Y')
```

```python
print(stores.dtypes)
```

```python
stores
```

```python
#Verify that StoreKey is unique
duplicates = stores[stores.duplicated(subset=['StoreKey'], keep=False)]
print(duplicates)
```

```python
#create a new data set
stores.to_csv("C:/Users/prem/Desktop/Ranjana/2.Data Spark Project/New Data Set/Stores.csv", index=False)
```
